```python
import random

question_bank = [
    {
        "question": "Which of the following is NOT a function of an Operating System?",
        "options": ["A) Process management","B) Memory management","C) Web designing","D) File management"],
        "answer": "C"
    },
    {
        "question": "A program in execution is called:",
        "options": ["A) Thread","B) Process","C) Job","D) Task"],
        "answer": "B"
    }
]

selected_questions = random.sample(question_bank, len(question_bank))

print("\n===== OPERATING SYSTEMS QUESTION PAPER =====\n")

for i, q in enumerate(selected_questions, 1):
    print(f"{i}. {q['question']}")
    for option in q["options"]:
        print("  ", option)
    print()

print("============================================")
```
